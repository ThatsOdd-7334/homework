````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def middlest_element(values):
    """Returns the middle element of a list.
    
    For odd-length lists, returns the exact middle.
    For even-length lists, returns the FIRST middle element.
    For empty lists, returns None.
    
    Args:
        values: A list of elements
        
    Returns:
        The middle element, or None if list is empty
    """
    if len(values) == 0:
        return None
    
    if len(values) % 2 == 1:  # odd length
        middle_index = len(values) // 2
        return values[middle_index]
    else:  # even length
        # Return the FIRST middle element
        middle_index = len(values) // 2 - 1
        return values[middle_index]


    


````

````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def mean(values):  #Calculating the mean of the values using the accumulator pattern
    """Calculates the population variance of a list of numbers."""

    total = 0

    for num in values:
        total = total + num
    return total / len(values)

def variance(values):
    """Calculates the population covariance between two series."""

    #Calculating the mean by using the mean() function

    x_bar = mean(values)

    #Initialize accumulator FOR the sum of squared differences
    sum_squared_diff = 0

    # Loop through each valuefor sum of squared differences
    for x in values:

        difference = x -x_bar # Calculate difference from mean 
        squared_diff = difference ** 2
        sum_squared_diff = sum_squared_diff + squared_diff

    # Calculate variance
    pop_var = sum_squared_diff / len(values)

    return pop_var
````

**Problem 8**

**Concept:** *Accumulator pattern with numeric result, list iteration, index-based loop.*

**Task:**

Write a function `covariance(xs, ys)` that computes the population covariance between two series:

$$
\text{covariance} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})
$$

Rules:

* If `xs` and `ys` are not the same length, print:
    
    `xs and ys must be the same length`
    
    and return `0`.

* Must use an index-based loop.

* Must use your `mean()` function.

* Then run the given test case and print exactly in the same format.

**Sample Output:**

```
x values: [6, 3, 7, 9, 5]
y values: [4, 1, 7, 5, 0]
covariance: 3.8
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def mean(values):
    """Calculates the mean of the list of numbers in values."""

    # Calculate the mean using accumulator pattern
    total = 0
    for num in values:
        total += num
    return total / len(values)

def covariance(xs, ys):
    """Calculates the population covariance between two series.
    
    Uses the formula: (1/n) * Œ£(x_i - xÃÑ)(y_i - »≥)
    Requires both lists to be the same length."""

    # Step 1: Check if lists are the same length
    if len(xs) != len(ys):
        print("xs and ys must be the same length")
        return 0
    
    # Step 2: Calculate means of both lists
    x_bar = mean(xs)
    y_bar = mean(ys)
    
    # Step 3: Initialize accumulator for sum of products
    sum_products = 0
    
    # Step 4: Use index-based loop to access corresponding elements
    for i in range(len(xs)):  # xs and ys have same length (checked above)
        # Calculate deviations from means
        x_dev = xs[i] - x_bar
        y_dev = ys[i] - y_bar
        
        # Multiply deviations and add to accumulator
        product = x_dev * y_dev
        sum_products += product
    
    # Step 5: Calculate covariance (average of products)
    cov = sum_products / len(xs)
    
    # Step 6: Return the result
    return cov
````



````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def polynomial_value(x, coefficients):
    """Evaluates a polynomial at given x value. Coefficients[i] is the coefficient for x^i term."""

    # Initialize accumulator for the result
    result = 0
    
    # Use index-based loop to access each coefficient
    for i in range(len(coefficients)):
        # Get the coefficient for x^i
        coeff = coefficients[i]
        
        # Calculate x^i and multiply by coefficient
        term = coeff * (x ** i)
        
        # Add the term to the accumulator
        result += term
    
    # Return the final polynomial value
    return result

````

````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def ascii_to_string(codes):

    """Converts a list of ASCII codes back to a string.
    
    Uses string accumulator pattern with concatenation."""
    
    # Initialize accumulator as an empty string
    result = ""
    
    # Loop through each ASCII code in the list
    for code in codes:
        # Convert ASCII code to character and concatenate
        result += chr(code)  # shorthand for result = result + chr(code)
    
    # Return the complete string
    return result


````

**Problem 12**

**Concept:** *Accumulator pattern to process a string, produce a string result.*

**Task:**

Write a function `cipher(s, n)` that applies a Caesar cipher shift to a plaintext string.

Rules:

* Assume lowercase letters only for shifting (`a` to `z`).

* Do not shift non-letters (spaces stay spaces, punctuation stays punctuation).

* Must wrap around the alphabet if shifting past `z` or before `a`.

* Must use an accumulator string.

* Then run the exact test cases shown and print in the same format.

**Sample Output:**

```
plaintext = abc, ciphered = def
plaintext = abc xyz, ciphered = def abc
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def cipher(s, n):
     """Applies a Caesar cipher shift to a string.
    Only shifts lowercase letters. Non-letters remain unchanged.
    Wraps around the alphabet using modulo 26."""


    # Initialize accumulator as an empty string
    result = ""
    
    # Loop through each character in the input string
    for char in s:
        # Check if the character is a lowercase letter
        if char.islower():
            # Convert character to number (0-25)
            # ord('a') = 97, so subtract 97 to get 0-25 range
            char_num = ord(char) - ord('a')
            
            # Apply Caesar shift (with wrap-around using modulo)
            shifted_num = (char_num + n) % 26
            
            # Convert back to character
            shifted_char = chr(shifted_num + ord('a'))
            
            # Add to result accumulator
            result += shifted_char
        else:
            # Non-letter characters (spaces, punctuation) stay the same
            result += char
    
    # Return the ciphered string
    return result
````

**Problem 13**

**Concept:** *More practice with loops.*

**Task:**

Write a function `print_all_decipherings(s)` that prints all 26 possible Caesar shifts of string `s`.

Rules:

* Use a definite loop for shifts `0` through `25`.

* For each shift `i`, compute `ciphered = cipher(s, i)` and print:

    `{i}, {ciphered}`

* Then run these two test cases:

    `s = "def"`

    `s = "oqfcgg hvs ibwjsfgs"`

**Sample Output:**

For `"def"` it must print 26 lines starting with `0, def` and ending with `25, cde`

For `"oqfcgg hvs ibwjsfgs"` it must print 26 lines. One of theses is the name of a Beatles song.



````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def cipher(s, n):
    """Prints all 26 possible Caesar shifts of a string.
    
    For each shift i from 0 to 25, prints "i, shifted_string.

    Uses the cipher function to apply to each shift."""
    
    # Reusing the cipher function from Problem 12
    result = ""
    for char in s:
        if char.islower():
            char_num = ord(char) - ord('a')
            shifted_num = (char_num + n) % 26
            shifted_char = chr(shifted_num + ord('a'))
            result += shifted_char
        else:
            result += char
    return result

def print_all_decipherings(s):
    # Loop through all possible shifts from 0 to 25
    for i in range(26):
        # Apply the cipher with shift i
        deciphered = cipher(s, i)
        # Print shift number and result in the required format
        print(f"{i}, {deciphered}")

````
